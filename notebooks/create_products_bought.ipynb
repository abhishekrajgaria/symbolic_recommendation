```python
from tqdm import tqdm

# Process the dataset to update the products_bought map
for product in tqdm(dataset_all_beauty["full"], desc="Processing dataset"):
    # Extract relevant values from the product object
    user_id = product['user_id']
    asin = product['asin']
    parent_asin = product['parent_asin']
    timestamp = product['timestamp']
    rating = product['rating']

    # Create a new product map with relevant key-values
    product_entry = {
        'asin': asin,
        'parent_asin': parent_asin,
        'timestamp': timestamp,
        'rating' : rating
    }

    # Check if the user_id exists in products_bought
    if user_id in products_bought:
        # Append the new product entry to the user's list
        products_bought[user_id].append(product_entry)
    else:
        # Create a new list for the user and add the product entry
        products_bought[user_id] = [product_entry]

# Get any top 10 key-value pairs
top_10 = islice(products_bought.items(), 10)

# Print the top 10 key-value pairs
for key, value in top_10:
    print(f"{key}: {value}")

del dataset_all_beauty
del top_10
```

```python
from pyspark.sql import SparkSession
from tqdm import tqdm




# Convert dictionary to a list of tuples
data_list = []

# Wrap `data.items()` with `tqdm` for a progress bar
for key, value in tqdm(products_bought.items(), desc="Processing Items", total=len(products_bought)):
    data_list.append((key, value))
    
del products_bought
```

```python
spark = SparkSession.builder \
    .appName("LargeDataProcessing") \
    .config("spark.executor.memory", "10g") \
    .config("spark.executor.instances", "5") \
    .config("spark.driver.memory", "10g") \
    .getOrCreate()

# Create a DataFrame
df = spark.createDataFrame(data_list, schema=["user_id", "products"])

# Save the DataFrame as Parquet
output_path = "user_products_parquet"
df.write.parquet(output_path)

print(f"Dictionary saved as Parquet at {output_path}")
```
